```python
from scipy import stats
import numpy as np
```

# What if the variances are different?

```python
sample_a = [6625,6000,5450,5200,5175,4900,4750,4500,3985,900,450,2800]
sample_b = [3900,3500,3450,3200,2980,2800,2500,2400,2200,1200,1150,1130]
```

```python
n = len(sample_a)
mean_a = np.mean(sample_a)
var_a = np.var(sample_a, ddof=1)
sd_a = var_a**(1/2)

mean_b = np.mean(sample_b)
var_b = np.var(sample_b, ddof=1)
sd_b = var_b**(1/2)
```

```python
t0 = (mean_a-mean_b)/(np.sqrt(var_a/n+var_b/n))
t0
```

```python
v = (var_a/n+var_b/n)**2 / ((var_a/n)**2/(n-1) + (var_b/n)**2/(n-1))
v
```

```python
t_dist = stats.t(v)
```

```python
1 - t_dist.cdf(t0)
```

```python
from scipy.stats import ttest_ind
ttest_ind(sample_a, sample_b, axis=0, equal_var=False)
```
